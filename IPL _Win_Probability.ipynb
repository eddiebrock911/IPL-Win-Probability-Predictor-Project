```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
# Old team names ko new team names se replace karna
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')
```

```python
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
# Sirf wahi rows rakho jisme teams valid hai
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df = match_df[match_df["dl_applied"] == 0]
```

```python
match_df = match_df[['match_id', 'city', 'winner', 'total_runs']]
```

```python
match_df
```

```python
delivery_df = match_df.merge(deliveries, on='match_id')
```

```python
delivery_df
```

```python
# Ensure 'total_runs_y' is numeric before cumsum
delivery_df['total_runs_y'] = pd.to_numeric(delivery_df['total_runs_y'], errors='coerce')
delivery_df['current_runs'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] +1 - delivery_df['current_runs']
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
# Missing values ko "0" se fill karna
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
# Agar player_dismissed "0" hai to wicket nahi, warna wicket hai (1)
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: 0 if x == "0" else 1)
# Int mein convert karna
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int)
# Cumulative wickets per match calculate karna
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum().values
# Bachi hui wickets = 10 - fallen wickets
delivery_df['wickets'] = 10 - wickets
# Check karne ke liye top rows
delivery_df.tail()
```

```python
# crr = (runs / overs) * 6
delivery_df['crr'] = (delivery_df['current_runs']*6) / (20 - (delivery_df['balls_left']))
```

```python
delivery_df.tail()
```

```python
# rrr = (runs left * 6) / balls left
delivery_df['rrr'] = (delivery_df['runs_left']*6) / delivery_df['balls_left']
```

```python
delivery_df.tail()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis=1)
```

```python
delivery_df.head(5)
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','total_runs_x','runs_left','balls_left','wickets','crr','rrr','result']]
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
final_df.isnull().sum()
```

```python
final_df.sample()
```

```python
# Downloading scikit_learn-1.7.2

from sklearn.model_selection import train_test_split

X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer(
    [
        ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
    ],
    remainder='passthrough'
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear',class_weight='balanced', max_iter=1000))
])
```

```python
X_train.describe()
```

```python
X_train.info()
```

```python
import numpy as np

# Replace inf/-inf with NaN, then drop rows with NaN
X_train = X_train.replace([np.inf, -np.inf], np.nan).dropna()
y_train = y_train.loc[X_train.index]

pipe.fit(X_train, y_train)
```

```python
# Replace inf/-inf with NaN, then drop rows with NaN in X_test
X_test = X_test.replace([np.inf, -np.inf], np.nan).dropna()
y_test = y_test.loc[X_test.index]
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
classification_report(y_test, y_pred)
```
